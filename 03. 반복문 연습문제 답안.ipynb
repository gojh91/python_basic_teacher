# 연습문제 11
(1) while 문과 break 문을 사용하여 사용자가 입력한 숫자만큼 반복하여 ‘hi’ 출력하기




```python
num = int(input("얼마나 인사를 할까요? "))
count = 0

while True:
    if num == count:
        break
    
    print("hi")
    
    count += 1
```

(2) for 문과 continue 문을 사용하여 사용자가 입력한 숫자까지의 짝수를 출력하기

```python
num = int(input("어디까지 짝수를 출력할까요?"))
for i in range(num+1):
    if i%2==1:
        continue
    print(i)
```
